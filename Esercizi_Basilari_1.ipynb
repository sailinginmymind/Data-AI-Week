# 10.WHILE

```python
# cosa sono i loop/cicli

x = ("milano", "roma", "napoli")
y = ("ciao")

print(x[0])
print(x[1])
print(x[2]) # farlo così è lungo e rindondante
# è come dire al cervello di fare un passo, un passo un passo
# per camminare gli dico "vai fino a quel punto, arrivato lì fai questo"
# il loop è un ciclo infinito di operazioni tutte uguali finchè una condizione è verificata

i = 0
while i < 6:
  print(i)
  i += 1 # se non si mette questo, va all'infinito
  # mi prendi i e mi aggiungi 1 alla i
  # ogni ciclo di i che si compie, si aggiunge +1 a i, fino a quando i < 6.
```

```python
# break, continue, else

#break serve a tagliare, a rompere e ad andare avanti

i = 0
while i < 6:
  print(i)
  i += 1
  if i == 3: # quando arrivi al 3, ti fermi
    break

print("ciao")

i = 0
while i < 6:
  i += 1
  if i == 3: # quando arrivi al 3, lo saltiamo quello che c'è dopo (print)
    continue
  print(i)

print("ciaoo")

i = 0
while i < 6:
  print(i)
  i += 1
else:
  print("ho finito") # non siamo obbligati a mettere else, serve per capire quando ha finito, per tenerne traccia

```

# 12.COLLEZIONE DI DATI

Introduzione teorica a collezioni di dati
Una collezione di dati è un insieme di più valori.

**Ordinata**: la collezione ha un ordine definito e l'aggiunta di elementi non incide;

es. citta = ("milano", "roma", "napoli") se aggiungo una città prima di milano, gli altri scalano.

**Indicizzato**: possiamo accedere agli elementi tramite indice (index)

**Modificabile**: possiamo aggiungere, cambiare e rimuovere elementi una volta creata la collezione;

**Immutabile**: non possiamo aggiungere, cambiare e rimuovere elementi;

**Permette duplicati**: possono esserci più elementi con lo stesso valore.


**List** -> collezioni ordinate e modificabili. Permettono duplicati.

**Tuple** -> collezioni ordinate ma immutabili. Permettono duplicati.

**Set** -> non ordinate, perciò non indicizzate. Non permettono duplicati.

**Dictionary** -> collezioni ordinate e modificabili. Non permettono duplicati.



# 13.LIST

```python
# una lista è una collezione di dati, ordinata, indicizzata, modificabile e permette duplicati

# si crea facendo:
x = [ "milano", "roma", "napoli"]
y = ("ciao", 100, False)

# len(), type() e list()

print(type(x)) # mi dice la classe, è una lista
print(len(x)) # mi dice quanti elementi ci sono
# la lista si crea o con le parentesi quadre [] o con list() e ci metti gli elementi dentro
print(list(x)) # mi crea una lista

# per accedere agli elementi della lista:
print(x[0])
print(x[1])

# range
print(x[1:3])

#modificare gli elementi
x = [ "milano", "roma", "napoli"]
x[0] = "brescia"

print(x)

#aggiungere -> aggiunge l'elemento in fondo
x = [ "milano", "roma", "napoli"]
x.append("brescia")
print(x)

# aggiungere alla posizione dove vuoi, non sostituisce, ma inserisce
x = [ "milano", "roma", "napoli"]
x.insert(1, "brescia")
print(x)

#extend -> x ha preso anche i valori di y, estende una lista con un'altra lista
x = [ "milano", "roma", "napoli"]
y = ("ciao", 100, False)
x.extend(y)

print(x)
print(y)

# rimuovere elementi
x = [ "milano", "roma", "napoli"]
x.remove("milano")
print(x)

# pop() -> rimuove l'index che decido io
x = ["milano", "roma", "napoli"]
x.pop(0)
print(x)

# del() -> cancelli lista
x = [ "milano", "roma", "napoli"]
del x[0]

print(x)

x = [ "milano", "roma", "napoli"]
x.clear() # svuota la lista
print(x)

# for in
x = [ "milano", "roma", "napoli"]
for citta in x:
  print(citta)

# per indice -> creami un range di tot numeri, tot numeri che è la lunghezza di x
x = [ "milano", "roma", "napoli"]
for i in range(len(x)):
  print(x[i])

# while
x = [ "milano", "roma", "napoli"]
i = 0
while i < len(x):
  print(x[i])
  i += 1

```

```python
# modificare l'ordine: sort, asc, desc, reverse

lista_citta = ["ancora", "milano", "roma", "bari", "napoli"]
lista_citta.sort() # ordine alfabetico
print(lista_citta)

lista_citta = ["ancora", "milano", "roma", "bari", "napoli"]
lista_citta.sort(reverse = True) # ordine alfabetico inverso
print(lista_citta)

# come copiare una lista -> fare una copia a se stante
x = ["ancora", "milano", "roma", "bari", "napoli"]
y = x.copy()
y[0] = "new york"
print(x)

# unire più liste insieme
x = ["ancora", "milano", "roma", "bari", "napoli"]
y = ["milano", "venezia"]

print(x+y) #o si fa così con la somma

x = ["ancora", "milano", "roma", "bari", "napoli"]
y = ["milano", "venezia"]
for citta in y:   #abbiamo preso ogni città di y e appesa in x
  x.append(citta)
print(x)
#si può fare anche con x.estend(y)

```

# 15.SET

```python
# collezione di dati non ordinate, non indicizzate, non modificabili e non permettono dubplicati

# creare set

x = {"milano", "roma", "napoli"}
z = {"milano", "34", "True"}

print(type(x))
print(len(x))

y = set(("milano", "roma", "napoli"))
print(type(y))

# non possiamo usare gli indici
# per accedere si fa con un loop

x = {"milano", "roma", "napoli"}
for citta in x:
  print(citta)

print("milano" in x) # ed è vero, esiste
print("milane" in x) # ed è falsa, non esiste in x

# aggiungere elementi (non con append ma con add)
x.add("venezia")
print(x)

# update -> uniamo un set ad un altro
x = {"milano", "roma", "napoli"}
y = {"pisa", "firenze"}

x.update(y)
print(x)

# rimuovere elementi con
# remove() di un elemento se siamo sicuri che esista
x = {"milano", "roma", "napoli"}
x.remove("milano")

# discard(), se l'elemento non esiste, non lo sapremo mai
x = {"milano", "roma", "napoli"}
x.discard("roma")

# pop() -> ci toglie l'ultimo elemento, non essendo ordinati elimina casualmente
x = {"milano", "roma", "napoli"}
x.pop()
print(x)

# clear() -> svuota il set
x = {"milano", "roma", "napoli"}
x.clear()
print(x)

# del() -> elimina tutto il set, ma ci da problemi perchè x non esiste

# intersection ci da solo gli elementi in comune e li mette in z
x = {"milano", "roma", "napoli"}
y = {"milano", "venezia", "firenze"}
z = x.intersection(y)
print(z)

# symmetric_difference ci tiene tutto tranne i duplicati
x = {"milano", "roma", "napoli"}
y = {"milano", "venezia", "firenze"}
z = x.symmetric_difference(y)
print(z)


```

# 17.FUNZIONI

```python
# cosa sono le funzioni
# un blocco di codice riutilizzabile al cui interno ci sono delle istruzioni che verranno eseguite ogni volta che richiamiamo la funzione

def fai_la_pasta():
    print("metti l'acqua")
    print("metti la pasta")
    print("metti il sale")

fai_la_pasta()
# se non avessi questo, ogni volta dovremmo copiare tutto.
# invece così scriviamo def una volta e poi basta richiamarlo
```

```python
# parametri
# è la variabile generica che usiamo nella definizione
# è il tipo di pasta

def fai_la_pasta(tipo_pasta):
    print("metti l'acqua")
    print("metti il sale")
    print("metti la " + tipo_pasta)

fai_la_pasta("rigatoni")
```

```python
# possiamo avere più parametri

def fai_la_pasta(tipo_pasta, metti_sugo):
    print("metti l'acqua")
    print("metti il sale")
    print("metti la " + tipo_pasta)
    if metti_sugo:
      print("prepara il sugo")

fai_la_pasta("rigatoni", True) # se metto true cado nell'if, senno no
```

```python
# arbitrary arguments
# argomenti che dobbiamo gestire quando non sappiamo quanti ne vogliamo

def fai_la_pasta(*opzioni): #con davanti asterisco
    print("metti l'acqua")
    print("metti il sale")
    print("metti la " + opzioni[0])
    if opzioni[1]:
      print("prepara il sugo")

fai_la_pasta("rigatoni", True)
```

```python
# keyword arguments
# si inserisce in fondo
def fai_la_pasta(tipo_pasta, sugo):
    print("metti l'acqua")
    print("metti il sale")
    print("metti la " + tipo_pasta)
    if sugo:
      print("prepara il sugo")

fai_la_pasta(sugo = True, tipo_pasta = "rigatoni")
```
