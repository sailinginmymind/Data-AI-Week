# 6.STRINGHE

```python
# - creare una stringa, parlare di apici e doppi apici
x = "ciao"
y = 'ciao'

# - mandare a schermo una stringa e variabile stringa
print(x)
print("Ciao")

# - stringhe multi riga """prova""" (tripli doppi o singoli apici)
x = """prova ciao
prova prova
 ciao
prova"""
print(x)

# trattare stringhe come array (prendere carattere, lenght, loop)
# l'array è una collezione di caratteri. E' l'indirizzo a cui troviamo il carattere
# è da vedere come x = prova quindi 0p 1r 2o 3v 4a
x
print(x[0]) # come print mi da "p" perchè sta nella prima posizione, cioè "0"
print(x[1]) # e così via :)
print(x[2])
print(x[3])
print(x[4])

# Per sapere la lunghezza della stringa si fa:

x = " Ciao, sono Alex come stai oggi? "
print(len(x)) # e dice che sono 31 caratteri, spazi e segni inclusi!!


# prendere parte di stringa

print(x[:7]) #con i due punti -> voglio qualcosa dall'inizio al settimo carattere (non compreso)

print(x[12:]) # due punti dopo -> parto dal fondo della stringa fino al 12esimo carattere


# Modificare una stringa con upper() lower() strip() split() e replace()

# upper() -> tutto il testo in maiuscolo, upper = uppercase = maiuscolo
print(x.upper())
# lower() -> tutto il testo in minuscolo
print(x.lower())
# strip() -> toglie gli spazi bianchi in fondo e davanti alla stringa
print(x.strip())
# replace() -> si sostituisce qualcosa nella stringa
print(x.replace("Alex","Stefano"))
# Oppure sostituire tutte le "o" con "w"
print(x.replace("o", "w"))


# Come concatenare le stringhe
x = "ciao sono"
y = "Alex"
print(x+y)

# Usare format() per combinare stringhe e numeri
# non posso scrivere -> prova = "ciao sono Alex e sono nato il" + 12
# posso scriverlo sono in parentesi graffe
# si può fare anche come variabile
x = 12
prova = "ciao sono Alex e sono nato il {}"
print(prova.format(12))
print(prova.format(x))

# Se ci sono più valori
prova = "ciao sono Alex e sono nato il {}, peso {} e sono alto {}."
print(prova.format(12, 80, 1.83))


# escape dei caratteri -> si usa backslash "\""
prova = "Ciao sono Alex e sono \"figo\""
print(prova)

# Problema più ricorrente con gli apici
prova = 'Ciao, sono Alex e penso all\'amore'
print(prova)

```

# 7.BOOLEAN

# 8.OPERAZIONI ARITMETICHE

```python
# Operatori aritmetici
# Operatori di Assegnamento
# Precedenza Operatori
# Metodi minx, max, abs, pow
```

```python
x = 15
y = 7

#Somma
print(x+y)
#Sottrazione
print(y-x)
#Moltiplicazione
print(x*y)
#Divisione
print(x/y)

#Modulo - ci da il resto della divisione (il primo numero del resto)
print(x%y)
#Potenza
print(x**y) # due asterischi
#Divisione (risultato arrontondato per difetto)
print(x//y)

# Prima "*"" e "/"" e poi "+"" e "-"

#Operatori di assegnamento

x = 5
x = x + 2 # se stesso + 2
y = 7
print(x)
# Scriverlo così è lungo, quindi si usano gli operatori di assegnamento

x = 15
x += 2 # segno + davanti all'uguale
y = 7  # il segno + davanti a tutte le altre operazioni
print(x)

# Metodi
x = min(5, 10, 25) # prende il numero più piccolo
print(x)

x = max(5, 10, 25) # prende il numero pià grande
print(x)

x = abs(-5) # toglie il segno che ha davanti, lascia solo il numero assolut
print(x)

x = pow(5,3) # la potenza, 5 alla terza
print(x)

```
