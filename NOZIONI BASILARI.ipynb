# 4.TIPI DI DATI

```python
# - ottenere il tipo di dato;
# - come viene asegnato il tipo
# - tipi di dati

# Il tipo viene assegnato in automatico appena si definisce un valore, tipo

x = 5
print(type(x))

# per capire il tipo del dato print(type(x)),
# a regola, ma noi siamo più furbi
# class int -> numero intero

x = 5.5
print(type(x))
# class float -> numero decimale

x = "ciao"
print(type(x))
# class str -> stringa di testo (può essere un link, un nome di persona, un papiro ecc)

x = True
y = False
print(type(x))
print(type(y))
#class bool -> valore booleano, vero o falso


# COLLEZIONI DI DATI

x = ["roma, milano, napoli"]
print(type(x))
#class list -> x è una lista, un contenitore di elementi

x = "roma", "milano", "napoli"
print(type(x))
# class tuple -> non serve ora

x = range(6)
print(type(x))
# class range -> non serve ora





```
