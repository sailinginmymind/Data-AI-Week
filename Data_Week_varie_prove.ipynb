```python
def somma(addendo_1, addendo_2):
    risultato = addendo_1 + addendo_2
    return risultato

 # si poteva anche scrivere return = addendo_1 + addendo_2 ed eliminare
 # dalla seconda riga "= addendo_1 + addendo_2"

```

```python
valore = somma(addendo_1 = 200, addendo_2 = 400)
print(valore)

#con print(valore) mi fa il calcolo
# inutile dire che potevo sostituire tutto con a, b, c
```

```python
valore = somma(200, 400)
print(valore)
```

```python
def é_pari(numero):
    return numero % 2 == 0

print(é_pari(7)) # Falso
print(é_pari(2)) # Vero

# Ricordati di:
# 1. Non usare lettere accentate (è,é ecc)
# 2. Se le usi, come sopra "è pari", assicurati che sia la stessa identica lettera.
```
